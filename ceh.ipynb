```python
import requests
import pandas as pd

# Base URL dari hasil copy Network Tab kalian (sesuaikan domain aslinya jika berbeda)
base_url = 'https://infopangan.jakarta.go.id/api2/v1/public/master-data/commodity/transaction/daily'

# Parameter query sesuai yang tertangkap di F12
params = {
    "commodityId": "8",              # ID untuk Cabai
    "startDate": "2020-01-01",        # Tarik mundur dari 2020
    "endDate": "2026-05-25"          # Sampai hari ini
}

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
}

response = requests.get(base_url, params=params, headers=headers)

if response.status_code == 200:
    data_json = response.json()
    
    # Asumsikan data list berada di dalam key 'data' atau langsung list mentah
    # Sesuaikan struktur json_data['data'] setelah kalian cek di browser
    df = pd.DataFrame(data_json['data']) 
    
    # Simpan langsung ke CSV siap pakai untuk XGBoost
    df.to_csv("data_target_cabai_jakarta.csv", index=False)
    print(f"Sukses! Berhasil mengunduh {len(df)} baris data.")
else:
    print(f"Gagal mengakses API. Status Code: {response.status_code}")
```
